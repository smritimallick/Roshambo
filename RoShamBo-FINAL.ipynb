```python
from tkinter import *
#The Python Imaging Library adds image processing capabilities 
## to your Python interpreter. 
from PIL import Image, ImageTk 
## The ImageTk module contains support to create and modify Tkinter BitmapImage
## and PhotoImage objects from PIL images.
from random import randint

root = Tk()
root.title("RoShamBo Game")
root.configure(background='lightblue')
root.geometry('1080x420')

#Resize the Image using resize method
# resized_image= img.resize((300,205), Image.ANTIALIAS)

rock = Image.open("Rock.png")
resized_rock= rock.resize((240,135), Image.ANTIALIAS)
image_rock = ImageTk.PhotoImage(resized_rock)

paper = Image.open("Paper.png")
resized_paper = paper.resize((240,135), Image.ANTIALIAS)
image_paper = ImageTk.PhotoImage(resized_paper)

scissors = Image.open("Scissors.png")
resized_scissors = scissors.resize((240,135), Image.ANTIALIAS)
image_scissors = ImageTk.PhotoImage(resized_scissors)

rock2 = Image.open("Rock2.png")
resized_rock2= rock2.resize((240,135), Image.ANTIALIAS)
image_rock2 = ImageTk.PhotoImage(resized_rock2)

paper2 = Image.open("Paper2.png")
resized_paper2 = paper2.resize((240,135), Image.ANTIALIAS)
image_paper2 = ImageTk.PhotoImage(resized_paper2)

scissors2 = Image.open("Scissors2.png")
resized_scissors2 = scissors2.resize((240,135), Image.ANTIALIAS)
image_scissors2 = ImageTk.PhotoImage(resized_scissors2)


label_player = Label(root, image = image_scissors2)
label_computer = Label(root, image = image_scissors)
label_player.grid(row = 2, column = 4)
label_computer.grid(row = 2, column = 0)

score_computer = Label(root, text=0, font= ('Verdana',30), fg='white', bg='lightblue')
#bg and fg for font and background color
score_player = Label(root, text=0, font= ('Verdana',30), fg= 'white', bg='lightblue')
score_player.grid(row = 2, column = 3)
score_computer.grid(row = 2, column = 1)

emptyline = Label(root, height = 3, bg='lightblue')
emptyline.grid(row=1)

indicator_player = Label(root, font= ('Verdana',20), fg= 'white', text = 'PLAYER', bg='lightblue')
indicator_computer = Label(root, font= ('Verdana',20), fg= 'white', text = 'COMPUTER', bg='lightblue')
indicator_player.grid(row = 0, column =3 )
indicator_computer.grid(row = 0, column = 1)

def updateMessage(a):
    final_message['text'] = a
    
def Computer_Update():
    final = int(score_computer['text'])
    final += 1
    score_computer['text'] = str(final)

def Player_Update():
    final = int(score_player['text'])
    final += 1
    score_player['text'] = str(final)
    
def winner_check(p,c):
    if p == c:
        updateMessage(" DRAW ")
    elif p == "rock" :
        if c == "paper":
            updateMessage(" COMPUTER WINS ")
            Computer_Update()
        else:
            updateMessage(" PLAYER WINS ")
            Player_Update()
                
    elif p == "paper" :
        if c == "scissors":
            updateMessage(" COMPUTER WINS ")
            Computer_Update()
        else:
            updateMessage(" PLAYER WINS ")
            Player_Update()
    elif p == "scissors" :
        if c == "rock":
            updateMessage(" COMPUTER WINS ")
            Computer_Update()
        else:
            updateMessage(" PLAYER WINS ")
            Player_Update()
    else:
        pass
    
to_select= ['rock', 'paper', 'scissors']

def choice_update(a):
    
    choice_computer = to_select[randint(0,2)]
    if choice_computer == 'rock':
        label_computer.configure(image= image_rock)
    elif choice_computer == 'paper':
        label_computer.configure(image= image_paper)
    else:
        label_computer.configure(image= image_scissors)
        
    if a == 'rock':
        label_player.configure(image= image_rock2)
    elif a == 'paper':
        label_player.configure(image= image_paper2)
    else:
        label_player.configure(image= image_scissors2)
        
    winner_check(a,choice_computer)
    
    
empty = Label(root, height = 3, bg='lightblue')
empty.grid(row=3)
final_message = Label(root, width = 15, height = 2, font= ('Verdana',20), fg= 'blue', bg='lightblue')
final_message.grid(row=5,column =2)

rock = Button(root, width = 12, height = 2, text = 'ROCK', font =('Verdana',12,'bold'), command = lambda: choice_update('rock')).grid(row=4, column =1)
paper = Button(root, width = 12, height = 2, text = 'PAPER', font =('Verdana',12,'bold'), command = lambda: choice_update('paper')).grid(row=4, column =2)
scissors = Button(root, width = 12, height = 2, text = 'SCISSORS', font =('Verdana',12,'bold'), command = lambda: choice_update('scissors')).grid(row=4, column =3)
#bg and fg for font and background color

root.mainloop()
```
